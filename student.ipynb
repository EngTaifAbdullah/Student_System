```python
# 1. Import psycopg2 library:
import psycopg2
from psycopg2 import OperationalError
```

```python
# 2. Connects to PostgreSQL server

def create_connection(db_user, db_password, db_host, db_port):
    connection = None
    try:
        connection = psycopg2.connect(
            user=db_user,
            password=db_password,
            host=db_host,
            port=db_port,
        )
        print("Connection to PostgreSQL server successful")
    except OperationalError as e:
        print(f"The error '{e}' occurred")
    return connection


connection = create_connection(
    "postgres", "TA123456at", "localhost", "5432"
)
```

```python
# 3. Create DB
def create_database(connection, query):
    connection.autocommit = True
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Database created successfully")
    except OperationalError as e:
        print(f"The error '{e}' occurred")


create_database_query = "CREATE DATABASE student_manage"
create_database(connection, create_database_query)
```

```python
# 4. 
def create_connection(db_name, db_user, db_password, db_host, db_port):
    connection = None
    try:
        connection = psycopg2.connect(
            database=db_name,
            user=db_user,
            password=db_password,
            host=db_host,
            port=db_port,
        )
        print("Connection to student_manage successful")
    except OperationalError as e:
        print(f"The error '{e}' occurred")
    return connection


connection = create_connection(
    "student_manage", "postgres", "TA123456at", "localhost", "5432"
)
```

```python
# 5. 
def execute_query(connection, query):
    connection.autocommit = True
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Query executed successfully")
    except OperationalError as e:
        print(f"The error '{e}' occurred")
```

```python
# 6. CREATE Student Table
create_students_table = """
CREATE TABLE IF NOT EXISTS students (
  student_id SERIAL PRIMARY KEY,
  student_name TEXT NOT NULL,
  email TEXT,
  age INTEGER
)
"""

execute_query(connection, create_students_table)
```

```python
# 7. CREATE Student Cource
create_courses_table = """
CREATE TABLE IF NOT EXISTS courses (
  course_id SERIAL PRIMARY KEY,
  course_name TEXT NOT NULL,
  course_code TEXT,
  credit_hours INTEGER
)
"""

execute_query(connection, create_courses_table)
```

```python
# 8. Instructor
create_instructors_table = """
CREATE TABLE IF NOT EXISTS instructors (
  instructor_id SERIAL PRIMARY KEY,
  instructor_name TEXT NOT NULL,
  email TEXT,
  department TEXT
)
"""

execute_query(connection, create_instructors_table)
```

```python
# 9. Create Enrollments
create_enrollments_table = """
CREATE TABLE IF NOT EXISTS enrollments (
  enrollment_id SERIAL PRIMARY KEY,
  student_id INTEGER REFERENCES students(student_id),
  course_id INTEGER REFERENCES courses(course_id),
  enrollment_date DATE
)
"""

execute_query(connection, create_enrollments_table)
```

```python
# 10. Add Students
students = [
    ("Taif", "Taif@email.com", 22),
    ("Rose", "roseee@email.com", 12),
    ("Lana", "Lanaa@email.com", 10)
]

student_records = ", ".join(["%s"] * len(students))

insert_query = f"""
INSERT INTO students (student_name, email, age)
VALUES {student_records}
"""

connection.autocommit = True
cursor = connection.cursor()
cursor.execute(insert_query, students)
```

```python
# 12.
from datetime import date

enrollments = [
    (1, 1, date.today()),
    (1, 2, date.today()),
    (2, 1, date.today()),
    (3, 3, date.today())
]

enrollment_records = ", ".join(["%s"] * len(enrollments))

insert_query = f"""
INSERT INTO enrollments (student_id, course_id, enrollment_date)
VALUES {enrollment_records}
"""

connection.autocommit = True
cursor = connection.cursor()
cursor.execute(insert_query, enrollments)
```

```python
# 13. 

def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except OperationalError as e:
        print(f"The error '{e}' occurred")
```

```python
# 14 . Read ALL Student
select_students = "SELECT * FROM students"
students = execute_read_query(connection, select_students)

for student in students:
    print(student)
```

```python
# 15. READ ALL Courses
select_courses = "SELECT * FROM courses"
courses = execute_read_query(connection, select_courses)

for course in courses:
    print(course)
```

```python
# 17. READ ALL ENROLL
select_enrollments = "SELECT * FROM enrollments"
enrollments = execute_read_query(connection, select_enrollments)

for enrollment in enrollments:
    print(enrollment)
```
